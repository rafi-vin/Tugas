# **Dasar-Dasar Pemrograman**

**1. Variabel dan Type Data**

Python adalah bahasa yang dynamic typed, yang berarti Anda tidak perlu mendeklarasikan tipe variabel. Tipe data dapat berupa int, float, string, list, dan sebagainya.


```
//Contoh Variabel dan Type Data

x = 5             # Integer
y = 3.14          # Float
name = "Python"   # String
is_active = True  # Boolean

```

**2. Operator dalam Python**

Python memiliki berbagai macam operator untuk melakukan operasi pada variabel dan nilai.

**a. Operator Aritmatika**

Digunakan untuk melakukan operasi matematika dasar.

Screenshot 2024-10-11 090814.png

```
//Contoh Operator Aritmatika

a = 10
b = 3
print(a + b)   # Penjumlahan, hasilnya 13
print(a - b)   # Pengurangan, hasilnya 7
print(a * b)   # Perkalian, hasilnya 30
print(a / b)   # Pembagian, hasilnya 3.333...
print(a % b)   # Modulus, hasilnya 1
print(a ** b)  # Eksponen, hasilnya 1000
print(a // b)  # Pembagian bulat, hasilnya 3

```

**b. Operator Perbandingan**

Digunakan untuk membandingkan dua nilai.

Screenshot 2024-10-11 090641.png

```
//Contoh Operator Perbandingan

a = 5
b = 10
print(a == b)  # False
print(a != b)  # True
print(a > b)   # False
print(a < b)   # True

```
**c. Operator Logika**

Digunakan untuk menggabungkan beberapa kondisi.

```
//Contoh Operator Logika

a = True
b = False
print(a and b)  # False
print(a or b)   # True
print(not a)    # False

```

**d. Operator Penugasan**

Digunakan untuk memberikan nilai ke variabel.

```
//Contoh Operator Penugasan

a = 10
a += 5  # Sama dengan a = a + 5
print(a)  # Hasilnya 15

```
**e. Operator Bitwise**

Operator bitwise digunakan untuk melakukan operasi pada level bit. Python mendukung berbagai operator bitwise yang bekerja dengan representasi biner dari angka.

```
//Contoh Operator Bitwise

a = 60  # 60 dalam biner = 0011 1100
b = 13  # 13 dalam biner = 0000 1101

print(a & b)  # AND, hasilnya 12 (0000 1100)
print(a | b)  # OR, hasilnya 61 (0011 1101)
print(a ^ b)  # XOR, hasilnya 49 (0011 0001)
print(~a)     # NOT, hasilnya -61 (komplemen biner)
print(a << 2) # Left shift, hasilnya 240 (1111 0000)
print(a >> 2) # Right shift, hasilnya 15 (0000 1111)

```

**f. Operator Ternary (Conditional Expression)**

Operator ternary atau dikenal juga sebagai conditional expression adalah cara singkat untuk menulis ekspresi kondisional seperti if-else dalam satu baris.

```
//Contoh Operator Ternary
nilai = 85
status = "Lulus" if nilai >= 75 else "Tidak Lulus"
print(status)  # Hasilnya "Lulus" karena nilai >= 75

```
